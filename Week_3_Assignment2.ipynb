```python
import pandas as pd
```

## Reading the dataframe we created in the first notebook

```python
postal_codes = pd.read_csv("Postal_Codes.csv")
```

## Reading the csv file provided in description

```python
coordinates = pd.read_csv("Geospatial_Coordinates.csv")
```

```python
coordinates
```

```python
postal_codes
```

#### As we can see , there is a column named Unnamed: 0 that just duplicates our index.

#### Dropping the Unnamed: 0 column (This column is just a mistake)

```python
postal_codes = postal_codes.drop("Unnamed: 0" , axis=1)
```

```python
postal_codes
```

## Joining two tables in one dataframe

```python
df = pd.merge(postal_codes , coordinates , how = "left")
```

```python
df
```

# Plotting

```python
import folium
```

```python
map_osm = folium.Map(location=[df["Latitude"][0],df["Longitude"][0]], zoom_start=10)
df.apply(lambda row:folium.CircleMarker(location=[row["Latitude"], row["Longitude"]], 
                                              radius=10)
                                             .add_to(map_osm), axis=1)
```

```python
map_osm
```

#### Here , we can see our dataframe works well.

## Now, we can save our dataframe to work on it elsewhere
